```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 

import pandas as pd
population_data = pd.read_csv('PopulationData.csv')
```

```python
#Get rid of first column it is useless
population_data = population_data.iloc[:, 1:len(population_data.axes[1])]
```

```python
population_data.head()
```

```python
#Get a random sample of 192 students
population_sample = population_data.sample(192, random_state=42)
```

```python
left_population = population_sample["Handed"].loc[population_sample['Handed'] == 'left']
left_population = left_population.count()

right_population = population_sample["Handed"].loc[population_sample['Handed'] == 'right']
right_population = right_population.count()
```

```python

objects = ('left', 'right')
y_pos = np.arange(len(objects))
distribution = [left_population,right_population]
 
plt.bar(y_pos, distribution, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of Handed')
plt.title('Handed Distribution')
plt.ylim(0,1000)
plt.show()
```

```python
left_population = population_data["Handed"].loc[population_data['Handed'] == 'left']
left_population = left_population.count()

right_population = population_data["Handed"].loc[population_data['Handed'] == 'right']
right_population = right_population.count()

objects = ('left', 'right')
y_pos = np.arange(len(objects))
distribution = [left_population,right_population]
 
plt.bar(y_pos, distribution, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Number of Handed')
plt.title('Handed Distribution')
plt.ylim(0,1000)
plt.show()
```

```python
#Compare the distribution between train_set and test_set
print(population_data.Handed.value_counts() / len(population_data))
print("\n")
print(population_sample.Handed.value_counts() / len(population_sample))
```

```python
import pandas as pd

computerData = pd.read_csv('computersData.csv')
#Get rid of first column it is useless
computerData = computerData.iloc[:, 1:len(computerData.axes[1])]
```

```python
computerData.describe(include='all')
```

```python
computerData.head()
```

```python
import random

random_sample = computerData.sample(n=450)
```

```python
random_sample.describe(include='all')
```

```python
import numpy as np
group1, group2, group3 = np.array_split(random_sample, 3)
```

```python
import matplotlib.pyplot as plt

plt.title("Random Sample age distributions")

box1 = plt.boxplot((group1['age'], 
                    group2['age'], 
                    group3['age']), vert = True, labels=( 'Group 1',
                                                                  'Group 2',
                                                                  'Group 3'))
```

```python
plt.show()
```

```python
group1['group'] = pd.Series(1, index=group1.index)
group2['group'] = pd.Series(2, index=group2.index)
group3['group'] = pd.Series(3, index=group3.index)
```

```python
group_stack = pd.DataFrame()
group_stack = group_stack.append(group1)
group_stack = group_stack.append(group2)
group_stack = group_stack.append(group3)
group_stack.describe()
```

```python
#axis = 0 is columns, axis = 1 is rows
pd.crosstab(group_stack.gender, group_stack.group, margins = True)
```
